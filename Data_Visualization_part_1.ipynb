```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the CSV file into a DataFrame
df = pd.read_csv('data.csv', encoding='latin1')

# Convert categorical variables to numerical representations
description_codes = pd.Categorical(df['Year']).codes
level_codes = pd.Categorical(df['Industry_code_NZSIOC']).codes

# Create a scatter plot
plt.scatter(description_codes, level_codes)
plt.xlabel('Year')
plt.ylabel('Industry_code_NZSIOC')
plt.title('Scatter Plot of Description and Level')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the CSV file into a DataFrame
df = pd.read_csv('data.csv', encoding='latin1')

# Create histograms for 'description' and 'level' columns
plt.figure(figsize=(10, 5))

# Histogram for 'Year'
plt.subplot(1, 2, 1)
plt.hist(df['Year'], bins=10, color='skyblue', edgecolor='black')
plt.xlabel('Year')
plt.ylabel('Frequency')
plt.title('Histogram of Description')

# Histogram for 'Industry_code_NZSIOC'
plt.subplot(1, 2, 2)
plt.hist(df['Industry_code_NZSIOC'], bins=10, color='salmon', edgecolor='black')
plt.xlabel('Industry_code_NZSIOC')
plt.ylabel('Frequency')
plt.title('Histogram of Level')

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Read the CSV file into a DataFrame
df = pd.read_csv('data.csv', encoding='latin1')
# Plot estimated density of values in a column
sns.kdeplot(df['Year'], shade=True)

# Add labels and title
plt.xlabel('Year')
plt.ylabel('Industry_code_NZSIOC')
plt.title('Density Estimate Plot')

# Show the plot
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the CSV file into a DataFrame
df = pd.read_csv('data.csv', encoding='latin1')

# Assuming 'column_name' is the column you want to create a pie chart for
# Replace 'column_name' with the actual column name from your DataFrame
column_name = 'Year'

# Calculate the counts for each category in the column
counts = df[column_name].value_counts()

# Plot a pie chart
counts.plot.pie(autopct='%1.1f%%', startangle=140)

# Add a title
plt.title('Pie Chart of ' + column_name)

# Display the plot
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

# Save the plot as an image
plt.savefig('pie_chart.png')

# Show the plot
plt.show()
```
